```python
ak=ak.drop(columns=['v1','Unnamed: 2','Unnamed: 3','Unnamed: 4'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
tf_idf_vect = CountVectorizer()
tf_idf_vect.fit(ak['v2'])
desc_matrix = tf_idf_vect.transform(ak["v2"])
```

```python
from sklearn.cluster import KMeans
```

```python
num_clusters = 3
km = KMeans(n_clusters=num_clusters)
km.fit(desc_matrix)
clusters = km.labels_.tolist()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sms = {'msg': ak["v2"].tolist(), 'Cluster': clusters}
frame = pd.DataFrame(sms, index = [clusters])
frame
```
